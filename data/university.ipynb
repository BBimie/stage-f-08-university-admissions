```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
traindf =pd.read_csv('https://raw.githubusercontent.com/HamoyeHQ/01-University-Admissions/master/data/Admission_Predict.csv')
```

```python
traindf
```

```python
traindf.columns = traindf.columns.str.strip().str.replace(" ",'_')
```

```python

traindf.info()
```

```python
sns.heatmap(traindf.corr(), annot=True) 
```

```python
sns.distplot(traindf.University_Rating, kde=False)
```

```python
traindf.groupby('University_Rating').Chance_of_Admit.hist(alpha=0.6)
```

```python
sns.stripplot(x='University_Rating', y='Chance_of_Admit', data=traindf, alpha=0.3, jitter=True)
```

```python
sns.swarmplot(x='TOEFL_Score', y='Chance_of_Admit', data=traindf)
```

```python
sns.swarmplot(x='GRE_Score', y='Chance_of_Admit', data=traindf)
```

```python
traindf.groupby('Research').Chance_of_Admit.hist(alpha=0.6)
```

```python
sns.swarmplot(x='Research', y='Chance_of_Admit', data=traindf)
```
